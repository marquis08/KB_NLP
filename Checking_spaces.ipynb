```python
import pandas as pd
import os
from tqdm import tqdm_notebook, tqdm

Data_dir="../KB_NLP/raw_data"
train_df = pd.read_csv(os.path.join(Data_dir,"train.csv"))
train_df['text'] = train_df['text'].astype(str) 
```

```python
import re
```

```python
train_df['length'] = train_df['text'].apply(lambda x: len(str(x).split(' ')))
train_df['spaces'] = train_df['text'].apply(lambda x: x.count(' '))
train_df['re_spaces'] = train_df['text'].apply(lambda x: len(re.findall('\s+', x)))
```

```python
train_nospace = train_df[train_df['length']==1]
```

```python
from konlpy.tag import Okt
from konlpy.tag import Hannanum
from konlpy.tag import Kkma
from konlpy.tag import Komoran
from konlpy.tag import Mecab
okt = Okt()
hanna = Hannanum()
kk = Kkma()
komo = Komoran()
#mecab = Mecab()
```

```python
okt_morphs = []
hanna_morphs = []
kk_morphs = []
komo_morphs = []
```

```python
train_nospace = train_nospace.reset_index()
```

```python
okt_morphs = []
for i in tqdm(range(len(train_nospace))):
    morph = okt.morphs(train_nospace['text'][i])
    okt_morphs.append(morph)
print(len(okt_morphs))
```

```python
for i in tqdm(range(len(train_nospace))):
    morph = hanna.morphs(train_nospace['text'][i])
    hanna_morphs.append(morph)
```

```python
for i in tqdm(range(len(train_nospace))):
    morph = komo.morphs(train_nospace['text'][i])
    komo_morphs.append(morph)
```

```python
train_nospace['okt'] = okt_morphs
train_nospace['hanna'] = hanna_morphs
train_nospace['komo'] = komo_morphs
```
